### Q1. What is Python, and why is it popular?

**Answer:**  
Python is a high-level, interpreted programming language that is easy to read and write. It is popular due to its simplicity, versatility, large community support, and wide applications in data science, web development, automation, and artificial intelligence.

### Q2. What is an interpreter in Python?

**Answer:**  
An interpreter in Python is a program that reads and executes code line-by-line, translating it into machine code at runtime. Pythonâ€™s interpreter allows for easier debugging and testing.

### Q3. What are pre-defined keywords in Python?

**Answer:**  
Pre-defined keywords are reserved words in Python that have special meanings and purposes, such as `if`, `else`, `for`, `while`, `True`, `False`, etc. They cannot be used as variable names.

### Q4. Can keywords be used as variable names?

**Answer:**  
No, keywords cannot be used as variable names because they are reserved for specific functions in the language syntax.

### Q5. What is mutability in Python?

**Answer:**  
Mutability refers to whether an object can be changed after it is created. Mutable objects like lists can be modified, while immutable objects like tuples and strings cannot be changed after creation.

### Q6. Why are lists mutable, but tuples are immutable?

**Answer:**  
Lists are designed to be dynamic and allow changes such as adding, removing, or modifying elements. Tuples, on the other hand, are meant to be fixed and secure, which is why they are immutable.

### Q7. What is the difference between `==` and `is` operators in Python?

**Answer:**  
`==` checks if the values of two variables are equal, while `is` checks if two variables refer to the same object in memory.

### Q8. What are logical operators in Python?

**Answer:**  
Logical operators are used to combine conditional statements. The three main logical operators in Python are `and`, `or`, and `not`.

### Q9. What is type casting in Python?

**Answer:**  
Type casting is the process of converting one data type into another, such as converting a string to an integer using `int()`, or an integer to a string using `str()`.

### Q10. What is the difference between implicit and explicit type casting?

**Answer:**  
- Implicit type casting is done automatically by Python when it converts a data type to another.
- Explicit type casting is done manually by the programmer using functions like `int()`, `float()`, or `str()`.

### Q11. What is the purpose of conditional statements in Python?

**Answer:**  
Conditional statements help in decision-making by executing different blocks of code based on certain conditions using `if`, `elif`, and `else` statements.

### Q12. Where does the `elif` statement work?

**Answer:**  
The `elif` (else if) statement is used when multiple conditions need to be checked. It is placed after an `if` statement and before the `else` block.

### Q13. What is the difference between `for` and `while` loops?

**Answer:**  
- `for` loops are used when the number of iterations is known.
- `while` loops are used when the condition needs to be checked repeatedly until it becomes false.

### Q14. Describe a scenario where a `while` loop is more suitable than a `for` loop.

**Answer:**  
A `while` loop is more suitable when the number of iterations is not known in advance. For example, reading input from a user until they type "exit".

```python
# Q6: Write a function to demonstrate the behavior of mutable and immutable arguments.
def demonstrate_mutable_immutable(mutable_arg, immutable_arg):
    """
    Demonstrates the behavior of mutable and immutable arguments in Python.

    Args:
        mutable_arg: A mutable object (e.g., a list).
        immutable_arg: An immutable object (e.g., an integer).
    """

    print("Original mutable argument:", mutable_arg)
    print("Original immutable argument:", immutable_arg)

    # Modify the mutable argument
    mutable_arg.append("modified")
    print("Mutable argument after modification within the function:", mutable_arg)

    # Modify the immutable argument (actually reassigns it)
    immutable_arg = immutable_arg + 1
    print("Immutable argument after modification within the function:", immutable_arg)

    print("Function finished.")


# Example usage:
my_list = [1, 2, 3]
my_int = 5

print("Before function call:")
print("my_list:", my_list)
print("my_int:", my_int)

demonstrate_mutable_immutable(my_list, my_int)

print("After function call:")
print("my_list:", my_list)  # my_list is modified!
print("my_int:", my_int)  # my_int is unchanged!
```

```python
# Q13: Write a Python program to find the sum of all even numbers between 1 and 50.
even_sum = 0
for i in range(1, 51):
    if i % 2 == 0:
        even_sum += i

print("Sum of even numbers from 1 to 50 is:", even_sum)
```

```python
# Q14: Write a program to reverse a string using a while loop.
original = "Python"
reversed_str = ""
index = len(original) - 1

while index >= 0:
    reversed_str += original[index]
    index -= 1

print("Original string:", original)
print("Reversed string:", reversed_str)
```

```python
# Q15: Write a Python program to calculate the factorial of a number provided by the user using a while loop.
num = 5  # Fixed input
factorial = 1
i = 1

while i <= num:
    factorial *= i
    i += 1

print("Factorial of", num, "is:", factorial)
```
